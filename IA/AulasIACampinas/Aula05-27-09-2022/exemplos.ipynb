```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression

#ceriando base de dados
dados = {'X':[5,15,25,35,45,55],
'Y' : [5,20,14,32,22,38]}

dados = pd.DataFrame(data = dados)

#separando os dados X é a variavel independente e Y é a variael dependente

X = dados['X'].values
Y = dados['Y'].values

#Função para transpor os valores da coordenada X
X = X.reshape(-1, 1)

#Criando a instancia para as classes regressão linear

ajuste1 = LinearRegression(fit_intercept= False) #y(x) = b1*x (b0 = 0)
ajuste2 = LinearRegression(fit_intercept= True) #y(x) = b0 + b1*x

#passamnd os dados para fazer os ajustes
ajuste1.fit(X, Y)
ajuste2.fit(X, Y)

#variaveis atribuidas para os valores dos minimos quadrados resultantes dos ajustes
R_sq1 = ajuste1.score(X, Y) 
R_sq2 = ajuste1.score(X, Y)

b0_ajuste1 = ajuste1.intercept_ #parametro b0 do ajuste 1
b1_ajuste1 = ajuste1.coef_ #parametro b1 do ajuste 1

b0_ajuste2 = ajuste2.intercept_ #parametro b0 do ajuste 2
b1_ajuste2 = ajuste2.coef_ #parametro b1 do ajuste 2

print("PARAMETROS OBTIDOS PARA O AJUSTE 1")
print("\ncoeficiente linear para o ajuste (b0): ", b0_ajuste1)
print("\ncoeficiente linear para o ajuste (b1): ", b1_ajuste1)
print("\nValor de R2 para o ajuste1: ", R_sq1)

print("PARAMETROS OBTIDOS PARA O AJUSTE 2")
print("\ncoeficiente linear para o ajuste (b0): ", b0_ajuste2)
print("\ncoeficiente linear para o ajuste (b1): ", b1_ajuste2)
print("\nValor de R2 para o ajuste1: ", R_sq2)

fig_size = plt.figure(figsize=(10, 6))
plot_ajuste1 = fig_size.add_subplot(121)
plot_ajuste1 = plt.scatter(X,Y,color = 'blue')
plot_ajuste1 = plt.plot(X, ajuste1.predict(X), color = 'red')
plot_ajuste1 = plt.title('Ajuste 1')
plot_ajuste1 = plt.xlabel('X')
plot_ajuste1 = plt.ylabel('Y')

plot_ajuste2 = fig_size.add_subplot(122)
plot_ajuste2 = plt.scatter(X,Y,color = 'blue')
plot_ajuste2 = plt.plot(X, ajuste1.predict(X), color = 'red')
plot_ajuste2 = plt.title('Ajuste 2')
plot_ajuste2 = plt.xlabel('X')
plot_ajuste2 = plt.ylabel('Y')
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import pandas as pd
from matplotlib import pyplot as plt
import statsmodels.formula.api as smf

x= [[2, 1], [5, 3], [15, 2], [25, 5], [35, 11], [45, 15], [55, 34], [60, 35],[64, 40], [70, 41], [78, 32], [83, 45], [90, 51], [95, 55], [100, 64], [104, 65]]
y= [4, 5, 20, 14, 32, 42, 38, 43, 36, 25, 50, 44, 52, 62, 68, 73]

ajuste1 = LinearRegression(fit_intercept=False)
ajuste2 = LinearRegression(fit_intercept=True)

ajuste1.fit(x,y)
ajuste2.fit(x,y)

R_sq1 = ajuste1.score(x,y)
R_sq2 = ajuste2.score(x,y)

b0_ajuste1 = ajuste1.intercept_
b1_ajuste1 = ajuste1.coef_[0]
b2_ajuste1 = ajuste1.coef_[1]

b0_ajuste2 = ajuste2.intercept_
b1_ajuste2 = ajuste2.coef_[0]
b2_ajuste2 = ajuste2.coef_[1]


print("\nPARAMETROS OBTIDOS PARA O AJUSTE 1")
print("\nCoeficiente linear para o ajuste1 (b0):", b0_ajuste1)
print("\nCoeficiente angular para o ajuste1(b1):", b1_ajuste1)
print("\nCoeficiente angular para o ajuste1(b2):", b2_ajuste1)
print("\nValor de r^2 para o ajuste1:", r_sq1)

print("\nPARAMETROS OBTIDOS PARA O AJUSTE 2")
print("\nCoeficiente linear para o ajuste1 (b0):", b0_ajuste2)
print("\nCoeficiente angular para o ajuste1(b1):", b1_ajuste2)
print("\nCoeficiente angular para o ajuste1(b2):", b2_ajuste2)
print("\nValor de r^2 para o ajuste1:", r_sq2)

df = pd.DataFrame(x, columns=['x1', 'x2'])
df['y'] = pd.Series(y)

#regressoa linear multipla usando o panda e statsmodels

df = pd.DataFrame(x, columns=['x1', 'x2'])
df['y'] = pd.Series(y)

ajuste_modelo = smf.ols(formula = 'y ~ x1 + x2', data = df)
equacao_ajustada = ajuste_modelo.fit()
equacao_ajustada.params

#preparando os dados para visualização
x_surf, y_surf = np.meshgrid(np.linspace(df.x1.min(),df.x1.max(), 100), np.linspace(df.x2.min(), df.x2.max(), 100))
onlyX = pd.DataFrame({'x1':x_surf.ravel(), 'x2':y_surf.ravel()})

fittedY = equacao_ajustada.predict(exog = onlyX)

fittedY = np.array(fittedY)

fig = plt.figure(dpi = 100)
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(df['x1'], df['x2'], df['y'], c = 'red', marker = 'o', alpha=0.5)
ax.plot_surface(x_surf, y_surf, fittedY.reshape(x_surf.shape), color ='blue', alpha = 0.3)
plt.show()
```
