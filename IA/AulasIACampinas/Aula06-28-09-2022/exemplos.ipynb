```python
import pandas as pd
from sklearn import tree 
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

df = pd.read_csv('../Dados/REGISTRO_SHOWS_DE_COMEDIA.csv', sep = ';')

print(df)

d = {'RJ' : 0, 'MG' : 1, 'SP' : 2}
df['estado'] = df['estado'].map(d)

d = {'N' : 0, 'S' : 1}
df['foi'] = df['foi'].map(d)

print(df)

caracteristicas = ['idade', 'carreira', 'pontos', 'estado']
X = df[caracteristicas]
Y = df['foi']

dtree = DecisionTreeClassifier()

dtree = dtree.fit(X,Y)

tree.plot_tree(dtree, feature_names = caracteristicas)

plt.show()

Idade = 38
Carreira = 12
Pontos = 11
Estado = 1

print("\n[0] ele não vai [1] ele vai")
print("\n",dtree.predict([[Idade,Carreira, Pontos, Estado]]))
```
