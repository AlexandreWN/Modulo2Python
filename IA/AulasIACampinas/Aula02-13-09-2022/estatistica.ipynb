Media, mediana e desvio padrão das idades para os personagens dos simpsons

```python
import pandas as pd
import numpy as np

grupo_1 = (1,8,10,38,39) #familia
grupo_2 = (8,10,39,45,49) #selecionados aleatorios

media_idade_grupo_1 = np.mean(grupo_1)
media_idade_grupo_2 = np.mean(grupo_2)

mediana_idade_grupo_1 = np.median(grupo_1)
mediana_idade_grupo_2 = np.median(grupo_2)

desvio_padrao_grupo_1 = np.std(grupo_1)
desvio_padrao_grupo_2 = np.std(grupo_2)

print("\nMédia para as idades do grupo 1 =", media_idade_grupo_1)
print("Mediana para as idades do grupo 1 =", mediana_idade_grupo_1)
print("Desvio padrão para as idades do grupo 1 =", desvio_padrao_grupo_1)

print("\nMédia para as idades do grupo 2 =", media_idade_grupo_2)
print("Mediana para as idades do grupo 2 =", mediana_idade_grupo_2)
print("Desvio padrão para as idades do grupo 2 =", desvio_padrao_grupo_2)
```

```python
import pandas as pd
import numpy as np

dados = {"grupo_1" : [1,8,10,38,39],
"grupo_2" : [8,10,39,45,49]}

dataframe = pd.DataFrame(data = dados)

media_idade_grupo_1 = dataframe['grupo_1'].mean()
mediana_idade_grupo_1 = dataframe['grupo_1'].median()
desvio_padrao_grupo_1 = dataframe['grupo_1'].std()

print("\nMédia para as idades do grupo 1 =", media_idade_grupo_1)
print("Mediana para as idades do grupo 1 =", mediana_idade_grupo_1)
print("Desvio padrão para as idades do grupo 1 =", desvio_padrao_grupo_1)
```

Boxplot (ferramenta que mostra a dispersão da mediana e pontos fora da curva) divide em quartis

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

dados = {"grupo_1" : [1,8,10,38,39],
"grupo_2" : [8,10,39,45,49]}

dataframe = pd.DataFrame(data = dados)

boxplot = dataframe[['grupo_1', 'grupo_2']].plot(kind='box', title='boxplot')
plt.show()


q3_grupo1, q1_grupo1 = np.percentile(dataframe['grupo_1'], [75, 25])

IQR_grupo1 = q3_grupo1 - q1_grupo1

print("Grupo1")
print("Idade máxima", max(dataframe['grupo_1']))
print("Idade minima", min(dataframe['grupo_1']))

print("q3", q3_grupo1)
print("q1", q1_grupo1)
print("IQR", IQR_grupo1)


q3_grupo2, q1_grupo2 = np.percentile(dataframe['grupo_2'], [75, 25])

IQR_grupo2 = q3_grupo2 - q1_grupo2

print("\nGrupo2")
print("Idade máxima", max(dataframe['grupo_2']))
print("Idade minima", min(dataframe['grupo_2']))

print("q3", q3_grupo2)
print("q1", q1_grupo2)
print("IQR", IQR_grupo2)
```

```python
from this import d
import numpy as np
from matplotlib import pyplot as pyplot
from scipy.stats import norm

dados = {'Densidade': [22.02, 23.83, 26.67, 25.38, 25.49, 23.50, 25.90, 24.89,
                       21.49, 22.67, 24.62, 24.18, 22.78, 22.56, 24.46, 23.79,
                       20.33, 21.67, 24.67, 22.45, 22.29, 21.95, 20.49, 21.81]}

media = np.mean(dados['Densidade'])
desvio = np.std(dados['Densidade'], ddof = 1)

plt.hist(dados['Densidade'])
plt.show()

dominio = np.linspace(np.min(dados['Densidade']), np.max(dados['Densidade']))
plt.plot(dominio, norm.pdf(dominio, media, desvio))
plt.show()
```
