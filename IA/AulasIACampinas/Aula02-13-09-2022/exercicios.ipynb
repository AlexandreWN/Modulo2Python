```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

dados = {'Mistura_1': [22.02, 23.83, 26.67, 25.38, 25.49, 23.50, 25.90, 24.89],
         'Mistura_2': [21.49, 22.67, 24.62, 24.18, 22.78, 22.56, 24.46, 23.79],
         'Mistura_3': [20.33, 21.67, 24.67, 22.45, 22.29, 21.95, 20.49, 21.81]}

dataframe = pd.DataFrame(data = dados)

media1 = dataframe['Mistura_1'].mean()
media2 = dataframe['Mistura_2'].mean()
media3 = dataframe['Mistura_3'].mean()

mediana1 = dataframe['Mistura_1'].median()
mediana2 = dataframe['Mistura_2'].median()
mediana3 = dataframe['Mistura_3'].median()

desvio1 = dataframe['Mistura_1'].std()
desvio2 = dataframe['Mistura_2'].std()
desvio3 = dataframe['Mistura_3'].std()

print("Média para mistura 1 =", media1)
print("Média para mistura 2 =", media2)
print("Média para mistura 3 =", media3)

print("\nMediana para mistura 1 =", mediana1)
print("Mediana para mistura 2 =", mediana2)
print("Mediana para mistura 3 =", mediana3)

print("\nDesvio padrão para mistura 1 =", desvio1)
print("Desvio padrão para mistura 2 =", desvio2)
print("Desvio padrão para mistura 3 =", desvio3)

boxplot = dataframe[['Mistura_1', 'Mistura_2', 'Mistura_3']].plot(kind='box', title='boxplot')
plt.show()

```

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import norm

dados={"Idade": [14, 17 ,18, 15, 15, 16, 17, 15, 16, 16, 15, 17, 15, 16,
16, 18, 18, 19, 17, 16, 17, 15, 16, 17, 17, 19, 20, 18,
17, 16, 15, 16, 16, 17, 18, 18, 17, 17, 15, 16, 16, 15]}

dataframe = pd.DataFrame(data = dados)

media = dataframe['Idade'].mean()
mediana = dataframe['Idade'].median()
desvio = dataframe['Idade'].std()
moda = dataframe['Idade'].mode()


print("Média é =", media)
print("\nMediana é =", mediana)
print("\nDesvio padrão é =", desvio)
print("\nModa é =", moda)

boxplot = dataframe[["Idade"]].plot(kind='box', title='boxplot')
plt.show()

plt.hist(dados['Idade'])
plt.show()

dominio = np.linspace(np.min(dados['Idade']), np.max(dados['Idade']))
plt.plot(dominio, norm.pdf(dominio, media, desvio))
plt.show()
```
